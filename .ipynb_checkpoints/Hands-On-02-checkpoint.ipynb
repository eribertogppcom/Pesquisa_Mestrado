# Prática 03: Uso do tcpdump para insperção de arquivos pcap.

Ao executarmos os códigos acima, prática 01 e prática 02, foram gerados arquivos **.pcap** e foram armazenados no diretório **ns3-Eclise**


Para que possamos analisar os mesmos com auxílio do **tcpdump** executamos os seguintes comandos:
    **tcpdump -nn -tt -r handsONp2-0-0.pcap**, 
    **tcpdump -nn -tt -r handsONp2-1-0.pcap** e 
    **tcpdump -nn -tt -r handsONp2-2-0.pcap**.

Aqui temos um detalhamento maior do envio e recebimento do pacote, podemos identificar o protocolo ARP sendo usado, tipo  de conexão, **UDP**, IP's, tamanho do pacote e tempo de entrega.
